### Create axes using Object oriented

#### create subplot 

```python
x = np.linspace(0,5,11)
y = x**2
fig,axes = plt.subplots(nrows=1,ncols=2) #create 3 rows and 2 column plot
plt.tight_layout() #smoothing, avoid overlap, and enlarge gaps between axes
#axes.plot(x,y)
```

```python
axes #axes are actually list of each axes in subplot
```

#### Plot value to subplot using iterate, this is smiliar with appending list.

```python
fig, axes = plt.subplots(nrows=1,ncols=2)
for curr_ax in axes:
    curr_ax.plot(x,y)
plt.tight_layout()
```

#### Draw axes using array index, draw only at correspond index's of array

```python
fig, axes = plt.subplots(nrows=1,ncols=2)
axes[0].plot(x,y)
```

```python
fig, axes = plt.subplots(nrows=1,ncols=2)
axes[0].plot(x,y,'blue')
axes[1].plot(x*2,y,'red')
axes[0].set_xlabel('xSub1')
axes[0].set_ylabel('ySub1')
axes[1].set_xlabel('xSub2')
axes[1].set_ylabel('ySub2')
plt.tight_layout()
```

### Figure Size and DPI

```python
fig = plt.figure(figsize=(4,6)) #width height size
ax = fig.add_axes([0,0,0.5,0.5])
ax.plot(x,y)
fig2 = plt.figure(figsize=(8,2))
ax2 = fig2.add_axes([0,0,0.5,0.5])
ax2.plot(y,x,'red')
```

```python
fig, axes = plt.subplots(nrows = 2,ncols=1,figsize=(8,2))
axes[0].plot(x,y)
axes[1].plot(y,x)
plt.tight_layout()
```

### Adding Legend

##### Tell which plot by label for each plot

```python
fig, ax = plt.subplots(nrows=1,ncols=2, figsize=(8,5))
ax[0].plot(x,y,label='Y', color='green') #green color
ax[0].plot(x,y**2,label='Y Squared', color='#00FFFF') #cyan color hex
ax[0].set_title('Test 1')
ax[0].set_xlabel('X')
ax[0].set_ylabel('Y')
ax[0].legend(loc=(0.1,0.8)) #loc = re positioning legend location (x,y)

ax[1].plot(x,y,label='Y')
ax[1].plot(y,x,label='Y Squared')
ax[1].set_title('Test 2')
ax[1].set_xlabel('X')
ax[1].set_ylabel('Y')
ax[1].legend(loc=(10)) #loc = re positioning legend location (encoded : 0-10)
```

### Basic color and line

##### arg list(in x, in y, color, line width, transparency, line style, marker, marker size)

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y,color='purple',linewidth = 3,alpha = 0.5
        ,linestyle='-.',marker="h", markersize = 25
        , markerfacecolor = 'cyan',markeredgecolor='black') 
ax.set_xlim([0,1]) #limit axis bound
ax.set_ylim([0,2])
```
