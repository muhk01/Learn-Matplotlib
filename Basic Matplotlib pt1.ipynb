## Basic Matplotlib Learn

```python
import numpy as np
```

```python
import pandas as pd
```

```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline #allow plot show in jupyter
```

### Draw Plot with Simple Array

```python
x = np.linspace(0,5,11)
y = x ** 2
```

```python
x
```

```python
y
```

### Create Plot and Customization

```python
plt.plot(x,y) #draw plot
```

```python
plt.plot(x,y,'red') #giving color
```

```python
plt.xlabel('Area X') #add label and title
plt.ylabel('Area Y')
plt.title('Array Linspace')
plt.plot(x,y,'green')
```

### Create Subplot (Multiple)

```python
plt.subplot(1,2,1) #arg details (row,col,plot num)
plt.plot(x,y,'red')

plt.subplot(1,2,2)
plt.plot(y,x,'black')
```

### Object oriented plot

```python
fig = plt.figure()
axes = fig.add_axes([0.1,0.1,0.8,0.8])
axes.set_xlabel('X Label')
axes.set_ylabel('Y Label')
axes.set_title('Array')
axes.plot(x,y)
```

#### Understanding add_axes method, smiliar like bbox input whereas [left,bottom,width,height]

```python
fig = plt.figure()
axes1 = fig.add_axes([0.1,0.1,0.8,0.8]) 
axes2 = fig.add_axes([0.2,0.5,0.4,0.3])
```

#### Try move inside axes into bottom corner of bigger one

```python
fig = plt.figure()
axes1 = fig.add_axes([0.1,0.1,0.8,0.8]) 
axes2 = fig.add_axes([0.5,0.1,0.4,0.3])
```

```python
fig = plt.figure()
axes1 = fig.add_axes([0.1,0.1,0.8,0.8]) 
axes2 = fig.add_axes([0.2,0.5,0.4,0.3])
axes1.plot(x,y)
axes1.set_title('Bigger')
axes1.set_xlabel('X label')
axes1.set_ylabel('Y label')

axes2.plot(y,x)
axes2.set_title('Smaller')
axes2.set_xlabel('X label')
axes2.set_ylabel('Y label')
```
